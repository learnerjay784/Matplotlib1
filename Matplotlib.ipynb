```python
#Session - 10
#Procedural/Functional oriented vs object oriented programming approach...
#1. Procedural/Functional Approach...
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(1,11)
plt.xlabel('N value')
plt.ylabel('Squ  are value of N') 
plt.plot(a,a**2,'o-b',ms = 10,mfc ='r')

#2 Object oriented programming approach...
'''In this approach, we have to create objects and on those objects
   we have to call corresponding methods to create a plot.'''
#Steps....
  # 1. Cretion of figure object...
a = np.arange(1,11)
b = a**2
fig = plt.figure()
# 2. Creation  of axes object...
axes = fig.add_axes([0.1,0.2,0.8,0.4]) #[left,bottom,width,height]
#plt.show()
#
# 3 plotting the graph and setting axes properties..
#------------------------------------------------------
axes.plot(a,b,'o-b',mfc ='y')
#axes.set_xlabel('N')
#axes.set_ylabel('Square of N')
#axes.set_title('Square Function')
#We can set all these values from a single method.....set()
axes.set(xlabel='N',ylabel = 'Square of N',title = 'Square Function',
         xlim =(1,5),ylim =(1,25))
axes.grid()
plt.show()
plt.savefig('squaregraph.jpg')
```

```python
#Burning Coil Animation...
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
fig = plt.figure
ax = plt.subplot(111)
ax.set_xlim(-100,100)
ax.set_ylim(-100,100)
x = []
y = []
line1, = ax.plot([],[],lw =2,color = 'y')
def init():
  print('init function called...')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('Burning Coil animation')
  global x,y 
  x,y = [],[]
  line1.set_data([],[])
  return line1 

def animate(i):
  print('Animate function called with i value:',i)
  t = 0.2*i 
  xdata = t*np.sin(t)
  ydata = t*np.cos(t)
  x.append(xdata)
  y.append(ydata)
  line1.set_data(x,y)

anm = FuncAnimation(fig,animate,init_func =init,frames = 500,interval = 10,repeat = True)
plt.show()

```
